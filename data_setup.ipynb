## Merging with Election Result Data
Now that we have cleaned the finance data, we will join the data with election result data.


### Cleaning of Elections Results Data:

```python
result_data = pd.read_csv("data/1976-2020-house.csv")
print(f"attribute types: {result_data.dtypes}")
print(f"num_rows: {result_data.shape[0]}")
print(f"contains nulls: \n{result_data.isna().any()}")
print(f"num duplicated rows: {result_data.duplicated().sum()}")
result_data.head(10)
```

```python
# Visualize entries that contain null values:
result_data[result_data['party'].isnull()].head(10)
```

```python
result_data[result_data['runoff'].isnull()].head(10)
```

```python
# Because the finance data only includes years >= 1990, drop all other years from election results data
result_data = result_data[result_data['year'] >= 1990]
print(f"num_rows: {result_data.shape[0]}")
result_data.head(10)
```

```python
# Drop write-in elections: outliers that may over-complicate results
result_data = result_data[result_data["writein"] == False]
print(f"num_rows: {result_data.shape[0]}")
result_data.head(10)
```

```python
# Drop special elections
result_data = result_data[result_data["special"] == False]
print(f"num_rows: {result_data.shape[0]}")
```

```python
# Drop anything that is not a general election
result_data = result_data[result_data["stage"] == "GEN"]
print(f"num_rows: {result_data.shape[0]}")
```

```python
# Visualize entries with mode != "TOTAL"
result_data_not_mode_total = result_data[result_data["mode"] != "TOTAL"]
print(f"num_rows: {result_data_not_mode_total.shape[0]}")
result_data_not_mode_total.head(10)
```

```python
# Visualize entries with fusion_ticket == True
result_data_fusion_ticket_true = result_data[result_data["fusion_ticket"] == True]
print(f"num_rows: {result_data_fusion_ticket_true.shape[0]}")
result_data_fusion_ticket_true.head(10)
```

```python
# Dropping rows with NaN in party
# A canidate must run under a specific party (even "Independent"). NaN values indicate that the specific candidate is 
# unknown (
result_data = result_data.dropna(subset="party")
print(f"party contains nulls: {result_data['party'].isna().any()}")
print(f"num_rows: {result_data.shape[0]}")
result_data.head(10)
```

```python
# How many fusion candidates do not have multiple entries for the same election? (Important for later cleaning)
columns = ["candidate", "year", "state", "district"]
num_duplicate_fusion_candidates = result_data[result_data["fusion_ticket"] == True][columns].duplicated().sum()
num_fusion_candidates = len(result_data[result_data["fusion_ticket"] == True])
print(f"num candidates without multiple entries: {num_fusion_candidates - num_duplicate_fusion_candidates}")
print(f"num duplicate entries: {num_duplicate_fusion_candidates}")
```

```python
# Candidates on fusion tickets (tickets where that candidate is listed under multiple parties) contain a entry with
# the number of votes they got per party they are running under
# The goal of the next couple of cells is to merge these entries in order to sum the number of votes they got and create
# a comma-separate value for the parties they are associated with
# This cell groups these candidates, sums their votes, and creates the corrrect value for the party
result_data_grouped_by_candidate = result_data[result_data["fusion_ticket"] == True].groupby(["year", "state_po", "candidate", "district"], as_index=False)
result_data_sum_votes = result_data_grouped_by_candidate.agg({
    'party': lambda x: ', '.join(x),
    'candidatevotes': "sum"
})
print(f"num_rows: {result_data_sum_votes.shape[0]}")
result_data_sum_votes.head(10)
```

```python
# We take the values from above and merge them into the original dataset
result_data = result_data.merge(right=result_data_sum_votes, 
                  how="left", 
                  on=["year", "state_po", "candidate", "district"])
print(f"num_rows: {result_data.shape[0]}")
result_data[result_data["fusion_ticket"] == True].head(10)
```

```python
# Set candidatevotes and party to appropriate values from the merge
result_data.loc[result_data["candidatevotes_y"].notna(), ["candidatevotes_x"]] = result_data[result_data["candidatevotes_y"].notna()]["candidatevotes_y"]
result_data.loc[result_data["party_y"].notna(), ["party_x"]] = result_data[result_data["party_y"].notna()]["party_y"]
print(f"num_rows: {result_data.shape[0]}")
result_data[result_data["fusion_ticket"]].head(10)
```

```python
result_data.drop(columns=["party_y", "candidatevotes_y"], inplace=True)
result_data.rename(columns={
    "candidatevotes_x": "candidatevotes",
    "party_x": "party",
}, inplace=True)
result_data.head(10)
```

```python
# There now exist duplicate values for many candidates. These should only be candidates that are part of
# fusion elections
# We will confirm that all duplicated values are from fusion tickets, and see how many there are
result_data[result_data.duplicated()]["fusion_ticket"].all()
print(f"num fusion candidates: {result_data['fusion_ticket'].sum()}")
print(f"num duplicates: {result_data.duplicated().sum()}")
```

```python
# Dropping duplicates
result_data.drop_duplicates(inplace=True)
print(f"num fusion candidates: {result_data['fusion_ticket'].sum()}")
print(f"num rows: {result_data.shape[0]}")
result_data.head(10)
```

```python
# Dropping unnecessary columns from result data
result_data.drop(
    columns=["state_po", "state_fips", "state_cen", "state_ic", "office", "runoff", "special", "writein", "mode", "unofficial", "version", "fusion_ticket"],
    inplace=True
)
result_data.head(10)
```

```python
# Format election result data names
result_data["split candidate"] = result_data["candidate"].str.split(" ")
result_data["Candidate Last Name"] = result_data["split candidate"].str[-1]
result_data["Candidate First Name"] = result_data["split candidate"].str[0]
result_data["Candidate Middle Name"] = result_data["split candidate"].str[1:-1]
result_data["Candidate Middle Name"] = result_data["Candidate Middle Name"].str.join(" ")
result_data.drop(columns="split candidate", inplace=True)
result_data.head(10)
```

```python
# Change result data column names
result_data.rename(
    columns={
        "candidatevotes": "Candidate votes",
        "totalvotes": "Total votes"
    },
    inplace=True
)
result_data.columns = map(str.capitalize, result_data.columns)
result_data.head(10)
```

We would like to calculate the election results for all House candidates
